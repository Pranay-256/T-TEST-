```python
import numpy as np
from scipy import stats
```

```python
sample = [172,174,168,169,171,173,175,170,169,172]
sample_mean=np.mean(sample)
sample_std=np.std(sample,ddof = 1)
n = len(sample)
mean_population = 170
```

```python
t_value = (sample_mean-mean_population)/(sample_std/np.sqrt(n))
t_value
```

```python
p_value=2*(1-stats.t.cdf(abs(t_value),df = n-1))
alpha = 0.05 # significance value
p_value
```

```python
if p_value<alpha:
    print("I will reject null hypothesis")
else:
    print("I will accept null hypothesis")
```
